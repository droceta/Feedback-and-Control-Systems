```python
import serial
import time

# Connect to Arduino on COM4 with baud rate 9600
arduino = serial.Serial(port='COM4', baudrate=9600, timeout=1)

# Give Arduino time to reset
time.sleep(2)
print("Connected to Arduino on COM4")

while True:
    command = input("Enter 1 to turn ON LED, 0 to turn OFF LED, x to quit: ")

    if command == '1':
        arduino.write(b'1')
        print("Sent: LED ON")
        print("Arduino says:", arduino.readline().decode().strip())
    elif command == '0':
        arduino.write(b'0')
        print("Sent: LED OFF")
        print("Arduino says:", arduino.readline().decode().strip())
    elif command == 'x':
        print("Exiting...")
        break
    else:
        print("Invalid command")

arduino.close()
```
