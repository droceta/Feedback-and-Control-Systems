```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1]       #numerator
den = [1, 2, 1]    # denominator s^2 + 2s + 1
G = ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zero = ctrl.zeros(G)
print("Zero:", zero)
```

```python
import matplotlib.pyplot as plt

ctrl.pzmap(G, grid=True)
plt.show()
```
